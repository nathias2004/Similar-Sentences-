```python
f = open("../data/phrases.csv","r")
Phrases = []
for phrase in f:
    Phrases.append(phrase.split("\n")[0])
Phrases = Phrases[1:]
```

```python
import re
def clean_str(string):
        '''
            Tokenization/string cleaning for all datasets except for SST.
            Original taken from https://github.com/yoonkim/CNN_sentence/blob/master/process_data
        '''

        string = re.sub(r"[^가-힣A-Za-z0-9(),!?:;.\'\`]", " ", string)
        string = re.sub(r"\'s", "s", string)
        string = re.sub(r"\'ve", " \'ve", string)
        string = re.sub(r"n\'t", " n\'t", string)
        string = re.sub(r"\'re", " \'re", string)
        string = re.sub(r"\'d", " \'d", string)
        string = re.sub(r"\'ll", " \'ll", string)
        string = re.sub(r"\.", " . ", string)
        string = re.sub(r",", " , ", string)
        string = re.sub(r":", " : ", string)
        string = re.sub(r";", " ; ", string)
        string = re.sub(r"!", " ! ", string)
        string = re.sub(r"\(", " ( ", string)
        string = re.sub(r"\)", " ) ", string)
        string = re.sub(r"\?", " ? ", string)
        string = re.sub(r"\s{2,}", " ", string)
        return string.strip()
    
```

```python
from nltk.tokenize import word_tokenize
Tokenized_Phrases = []
for phrase in Phrases:
    Tokenized_Phrases.append(word_tokenize(clean_str(phrase).lower()))
print(Tokenized_Phrases)
```

```python
from gensim.models import Word2Vec
# train model
model = Word2Vec(Tokenized_Phrases, min_count=1)
model.save('model.bin')
```

## Sentence Embeddings

* Calculating similarity between English phrases (phrases.csv)
  * Write the pseudo code (or high level description) on how to calculate similarities of phrases in the file phrases.csv using the above code that you have written.
  * Write the code in Python 3.x to perform the calculation.
  * For each phrase list the top 3 most similar other phrases in descending order of similarity

```python
#To get Sentence embeddings i can use simple averaging of the wordtovec of the sentence or else i can take weighted average
#In this i am using simple averaging
from scipy import spatial
def SentenceVector(phrase,model):
    phrase = clean_str(phrase).lower()
    Tokens = word_tokenize(phrase)
    vec = np.array(model.wv[Tokens[0]])
    for token in Tokens[1:]:
        vec = np.add(vec,np.array(model.wv[token]))
    return vec
def Top_Similar_Sentences(phrase,phrases,model):
    sen1_vec = SentenceVector(phrase,model)
    Scores = []
    for sen in phrases:
        sen2_vec = SentenceVector(sen,model)
        similarity_score = spatial.distance.cosine(sen1_vec,sen2_vec)
        Scores.append([similarity_score,sen])
    Scores = sorted(Scores, key=lambda x: x[0])
    print(Scores[1:4])
        
for phrase in Phrases:
    print(phrase)
    Top_Similar_Sentences(phrase,Phrases,model)
    
```
